```python
import cv2
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
img1 = cv2.imread("data/white bg.JPG")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("data/blue sky.jpeg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img1)
```

```python
plt.imshow(img2)
```

Blend and paste image

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))

blended = cv2.addWeighted(img1, 0.8, img2, 0.8, 0)

plt.imshow(blended)
```

Apply a binary threshold onto the image.

```python
img = cv2.imread('data/white bg.JPG', 0)
ret, thresh1 = cv2.threshold(img, 127,255, cv2.THRESH_BINARY)
```

```python
display_img(thresh1, cmap='gray')
```

Use a kernel to blur the img

```python
import numpy as np
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
kernel
```

```python
img = cv2.imread('data/white bg.JPG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(img, -1, kernel)
display_img(result)
```

Sobel Gradient

```python
img = cv2.imread('data/white bg.JPG', 0)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=9)
display_img(sobelx, 'gray')
```

```python
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=9)
display_img(sobely, 'gray')
```

plot the color histogram for R B G

```python
img = cv2.imread('data/blue sky.jpeg')

color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None, [256], [0,256])
    plt.plot(histr, color=col)
plt.title('HISTO')
```
