```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read an image and show it

```python
img = cv2.imread("/Users/tianyangliu/Documents/me/ID photo/white bg.JPG")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

Flip the image and show it

```python
new_img = img_rgb.copy()
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

Draw an empty red rectangle on the face

```python
pt1 = (750, 750) # top left
pt2 = (1600, 1600) # bottom right

cv2.rectangle(img_rgb, pt1=pt1,pt2=pt2, color=(255,0,0), thickness=10)

plt.imshow(img_rgb)
```

Create an event to draw circle on the picture

```python
def draw_circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100, (0,0,255), thickness=15)

img = cv2.imread("/Users/tianyangliu/Documents/me/ID photo/white bg.JPG")

cv2.namedWindow(winname='id photo')

cv2.setMouseCallback('id photo', draw_circle)

while True:

    cv2.imshow('id photo', img)

    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
